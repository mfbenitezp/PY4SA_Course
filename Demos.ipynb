One of the most traditional practice of Loops and Conditionals is the game of Rock, Paper, Scissors.

````python
import random
print("Welcome to ROCK PAPER SCISSORS - GG3209")

while True:

    print('....') 
    print("Enter your choice \n 1 - Rock \n 2 - Paper \n 3 - Scissors \n") 
    # take the input from user
     
    choice=int(input("Enter your choice :"))
    if choice == 1:
        choice_name= 'Rock'
    elif choice == 2:
        choice_name= 'Paper'
    else:
        choice_name= 'Scissors'
    
    # Print user choice
    print('The User choice is \n', choice_name)
    print('....')
    print('Now its Computers Turn...it will take literally one second')
    print('....')

    # get a random number from 1 to 3. 
    comp_choice = random.randint(1,3)
    
    # Take the random number and assign that to an game option.
    if comp_choice == 1:
        comp_choice_name = 'rocK'
    elif comp_choice == 2:
        comp_choice_name = 'papeR'
    else:
        comp_choice_name = 'scissoR'
    

    # Print Computer choice
    print("The Computer has chosen \n", comp_choice_name)
    print('....')
    print("So, the fight is ", choice_name,'Vs',comp_choice_name)
    print('....')

    # Game options
    if choice == comp_choice: # Draw
        result="DRAW"
    if (choice==1 and comp_choice==2): #Paper wins 
        result='papeR'
    elif (choice==2 and comp_choice==1): 
        result='Paper'
    if (choice==1 and comp_choice==3): # Rock Wins
        result='Rock'
    elif (choice==3 and comp_choice==1):
        result='rocK'
    if (choice==2 and comp_choice==3): # Scissor Wins
        result='scissoR'
    elif (choice==3 and comp_choice==2):
        result='Rock'

    # Printing who wins... either user or computer wins or draw
    if result == 'DRAW':
        print("<== Its a tie, nobody wins ==>")
    elif result == choice_name:
        print("<== User wins ==>")
    else:
        print("<== Computer wins ==>")
    
    # Break the infinitive while 
    print("Do you want to play again? (Y/N)")
    # If user input N then condition is True
    ans = input()
    if ans =='N':
        break
# we print thanks for playing
print("Thanks for playing")
````

### Ways to create a Pandas DataFrame

| Create DataFrame from: | Code                                                                                           |
|-----------------------|------------------------------------------------------------------------------------------------|
| Lists of lists        | ```pd.DataFrame([['Tom', 7], ['Mike', 15], ['Tiffany', 3]])```                                    |
| ndarray               | ```pd.DataFrame(np.array([['Tom', 7], ['Mike', 15], ['Tiffany', 3]]))```                             |
| Dictionary            | ```pd.DataFrame({"Name": ['Tom', 'Mike', 'Tiffany'], "Number": [7, 15, 3]})```                       |
| List of tuples        | ```pd.DataFrame(zip(['Tom', 'Mike', 'Tiffany'], [7, 15, 3]))```                                      |
| Series                | ```pd.DataFrame({"Name": pd.Series(['Tom', 'Mike', 'Tiffany']), "Number": pd.Series([7, 15, 3])})``` |

````python
pd_1 = pd.DataFrame([['Tom', 7], ['Mike', 15], ['Tiffany', 3]])
pd_2 = pd.DataFrame(np.array([['Tom', 7], ['Mike', 15], ['Tiffany', 3]]))
pd_3 = pd.DataFrame({"Name": ['Tom', 'Mike', 'Tiffany'], "Number": [7, 15, 3]})
pd_4 = pd.DataFrame(zip(['Tom', 'Mike', 'Tiffany'], [7, 15, 3]))
pd_5 = pd.DataFrame({"Name": pd.Series(['Tom', 'Mike', 'Tiffany']), "Number": pd.Series([7, 15, 3])})
pd_1
````

### Indexing with ```.loc``` and ```.iloc```

Pandas created the methods **.loc[]** and **.iloc[]** as more flexible alternatives for accessing data from a dataframe. 

Use ```df.iloc[]``` for indexing with integers and ```df.loc[]``` for indexing with labels.

These are typically the recommended methods of indexing in Pandas

````python
df.iloc[0]  # returns a series
````

````python
df.iloc[0:2] # slicing returns a dataframe
````

````python
df.iloc[2, 1] # returns the indexed object
````

````python
df.iloc[[0, 1], [1, 2]]  # returns a dataframe
````

````python
df.loc[:, 'Name']  # returns a series
````

````python
df.loc[:, 'Name':'Language'] # returns a dataframe
````

````python
df[df['Courses'] > 5]
````

````python
df[df['Name'] == "Tom"]
````
