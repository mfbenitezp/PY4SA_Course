## 1.5 Nested if conditions

In some cases it might be useful to have nested if statements, meaning that you have another layer of conditions after the first condition resolves to True. The following example has some issue, adress the issue and check the nested if conditions:

````python
season = "Winter"
 temperature = 10

 if season == "Winter":

     if temperature > 7:
         print("No need for winter jacket!")

     else:
         print("It might be cold! Wear a proper jacket!")

elif season == "Summer":

     if temperature > 20:
         print("It's warm! Time to wear shorts!")

     else:
         print("Well this is Finland, better wear long trousers!")
else:
     print("Check the weather forecast!")
````
